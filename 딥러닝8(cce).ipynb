```python
import tensorflow as tf

from tensorflow.keras.losses import SparseCategoricalCrossentropy

batch_size ,n_class = 16,5

prediction = tf.random.uniform(shape=(batch_size,n_class),minval=0,maxval=1,dtype=tf.float32)

pred_sum = tf.reshape(tf.reduce_sum(prediction,axis=1), (-1,1))
prediction = prediction/pred_sum

labels = tf.random.uniform(shape=(batch_size,),
                           minval=0,maxval=n_class,
                           dtype=tf.int32)

loss_object = SparseCategoricalCrossentropy()
loss = loss_object(labels,prediction)

print(loss.numpy())

ce = 0
for label,prediction in zip(labels,prediction):
    ce += -tf.math.log(prediction[label])
ce/= batch_size
print(ce.numpy())
```

```python
import tensorflow as tf

from tensorflow.keras.layers import Dense

from tensorflow.keras.losses import SparseCategoricalCrossentropy
N ,n_feature = 100,2
n_class = 5

X = tf.zeros(shape=(0,n_feature))
Y = tf.zeros(shape=(0,),dtype=tf.int32)

for class_idx in range(n_class):
    center = tf.random.uniform(minval=-15,maxval=15,shape=(2,))

    x1 = center[0] + tf.random.normal(shape=(N,1))
    x2 = center[1] + tf.random.normal(shape=(N,1))

    x = tf.concat((x1,x2),axis=1)
    y = class_idx*tf.ones(shape=(N,1),dtype=tf.int32)

    X = tf.concat((X,x),axis=0)
    Y = tf.concat((Y,y),axis=0) 

dataset = tf.data.Dataset.from_tensor_slices((X,Y))
datasef = dataset.batch(batch_size)

model = Dense( units=n_class, activation='softmax')
loss_object = SparseCategoricalCrossentropy()

for x,y in dataset:
    predictions = model(x)
    loss = loss_object(y,prediction)
    print(loss.numpy())

```

```python
import tensorflow as tf

from tensorflow.keras.losses import CategoricalCrossentropy

batch_size ,n_class = 16,5

prediction = tf.random.uniform(shape=(batch_size,n_class),minval=0,maxval=1,dtype=tf.float32)

pred_sum = tf.reshape(tf.reduce_sum(prediction,axis=1), (-1,1))
prediction = prediction/pred_sum

labels = tf.random.uniform(shape=(batch_size,),
                           minval=0,maxval=n_class,
                           dtype=tf.int32)

labels = tf.one_hot(labels,n_class)

loss_object = CategoricalCrossentropy()
loss = loss_object(labels,prediction)

tmp= tf.reduce_mean(tf.reduce_sum(-labels*tf.math.log(prediction),axis=1))
print(tmp.numpy())
print(loss.numpy())
```
