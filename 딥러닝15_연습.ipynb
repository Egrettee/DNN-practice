```python
import tensorflow as tf

from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten

N,n_H,n_W,n_C = 32,28,28,3
k_size=3
n_conv_filter=5
pool_size,pool_strides = 2,2
batch_size = 32

x = tf.random.uniform(shape=(N,n_H,n_W,n_C))

conv1 = Conv2D(filters=n_conv_filter,kernel_size=k_size,activation='relu',padding='same')

conv_pool1 = MaxPooling2D(pool_size = pool_size,strides=pool_strides)

conv2 = Conv2D(filters=n_conv_filter,kernel_size=k_size,activation='relu',padding='same')

conv_pool2 = MaxPooling2D(pool_size = pool_size,strides=pool_strides)

flatten = Flatten()

x = conv1(x)
x = conv_pool1(x)

x = conv2(x)
x = conv_pool2(x)

x = flatten(x)
print(x.shape)
```

```python
from tensorflow.keras.layers import Dense

n_neurons= [50,25,8]

dense1 = Dense(units=n_neurons[0],activation='relu')
dense2 = Dense(units=n_neurons[1],activation='relu')
dense3 = Dense(units=n_neurons[2],activation='softmax')

x = dense1(x)
x = dense2(x)
x = dense3(x)

W,B = dense3.get_weights()
print(W.shape,B.shape)
print(x.shape)
```

```python
from tensorflow.keras.losses import CategoricalCrossentropy

y = tf.random.uniform(minval=0,maxval=8,shape=(32,),dtype=tf.int32)

y = tf.one_hot(y,depth=8)

loss_object = CategoricalCrossentropy()
loss = loss_object(y,x)
print(loss)
print(loss.shape)
```
