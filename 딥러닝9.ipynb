```python
import tensorflow as tf

from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Model

class TestModel(Model):
    def __init__(self):
        super(TestModel,self).__init__()

        self.dense1 = Dense(units=5,activation='sigmoid')
        self.dense2 = Dense(units=3,activation='sigmoid')

    def call(self,x):
        x = self.dense1(x)
        x = self.dense2(x)

        return x

x = tf.random.normal(shape=(3,5))
model =  TestModel()
Y = model(x)
print(Y.numpy())
```

```python
import tensorflow as tf

from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Model

from tensorflow.math import exp,maximum

activation = 'tanh'

x = tf.random.normal(shape=(5,8))
dense = Dense(units=1,activation=activation)

y_tf = dense(x)
W,B = dense.get_weights()

y_man = tf.linalg.matmul(x,W) + B

if activation =='sigmoid':
    y_man =  1/(1+exp(-y_man))

elif activation =='relu':
    y_man = maximum(y_man,0)

elif activation == 'tanh':
    y_man = (exp(y_man)-exp(-y_man))/(exp(y_man)+exp(-y_man))

print(y_tf.numpy())
print(y_man.numpy())
```
